```python
# Setup WebDriver for Chrome (no Edge)
service = ChromeService(ChromeDriverManager().install())
options = ChromeOptions()

driver = webdriver.Chrome(service=service, options=options)

Title = []
Price = []
Table = []
Built_up_area =[]
Furnishing = []
Available_from = []
Lease_type = []
Locality = []
Bathrooms = []
Balcony = []
Carpet_area = []
Gas_Pipeline = []
Gate_Community = []
Floor_number=[]
Total_floors=[]
for link in all_house_links[:400]:
    url_ = link
    # print(url_)
    driver.get(url_)
    time.sleep(3)
    
    soup_ = BeautifulSoup(driver.page_source, "html.parser")
    
    # House Title
    title = soup_.find("div", class_="css-1hidc9c")
    if title:
        Title.append(title.text)
    else:
        Title.append(np.nan)

    # House Price
    price = soup_.find("span",class_="css-gg3jj9")
    if price:
        Price.append(price.text)
    else:
        Price.append(np.nan)
        
    # ALl Information in Table
    table=soup_.find("table",class_="_9s1ule T_block")
    if table:
        Table.append(table.text)
    else:
        Table.append(np.nan)
    text_block = soup_.get_text(" ", strip=True)
    
    pattern = r"(\d{2,6})\W*sq\.ft"
    built_up_area = re.findall(pattern, text_block, flags=re.IGNORECASE)
    
    if built_up_area:
        Built_up_area.append(built_up_area[0])
    else:
        Built_up_area.append(np.nan)
    # Furnishing Extraction
    furn_pattern = r"Furnishing\s*(Semi Furnished|Fully Furnished|Unfurnished|Partially Furnished|Furnished)"
    furnishing = re.findall(furn_pattern, text_block, flags=re.IGNORECASE)
    
    if furnishing:
        Furnishing.append(furnishing[0].title())
    else:
        Furnishing.append(np.nan)
    # =============== Available From Extraction ===============
    avail_pattern = r"Available from\s*(.+?)(?=Floor number|Lease type|Age of property|Parking|Main entrance|Gas Pipeline|Gate Community|Carpet area|Bathrooms|Balcony|Price|$)"
    
    available = re.findall(avail_pattern, text_block, flags=re.IGNORECASE)
    
    if available:
        avail_clean = available[0].strip()
        avail_clean = re.sub(r"\s+", " ", avail_clean)
        Available_from.append(avail_clean)
    else:
        Available_from.append(np.nan)

     # =============== Lease Type Extraction ===============
    lease_pattern = r"Lease type\s*([A-Za-z /]+?)(?=Age|Parking|Main entrance|Gas|Gate|Carpet|Bath|Balcony|Price|Floor|$)"
    
    lease = re.findall(lease_pattern, text_block, flags=re.IGNORECASE)
    
    if lease:
        lease_clean = lease[0].strip()
        lease_clean = re.sub(r"\s+", " ", lease_clean)
        Lease_type.append(lease_clean)
    else:
        Lease_type.append(np.nan)
   # Locality extraction (simple last 2 parts)
    locality = soup_.find("div", class_="_h31y44 _csbfng _c8exct _g38jkm _7lc65e T_addressStyle")
    if locality:
        loc_text = locality.text.strip()
    
        # Split on commas
        parts = [p.strip() for p in loc_text.split(",")]
    
        if len(parts) >= 2:
            # last two parts only
            simple_loc = ", ".join(parts[-2:])
            Locality.append(simple_loc)
        else:
            Locality.append(loc_text)
    else:
        Locality.append(np.nan)
    # =============== Carpet Area Extraction ===============
    carpet_pattern = r"Carpet\s*area\s*([0-9,]+)\s*sq\.ft"
    carpet = re.findall(carpet_pattern, text_block, flags=re.IGNORECASE)
    
    if carpet:
        carpet_clean = carpet[0].replace(",", "")
        Carpet_area.append(carpet_clean)
    else:
        Carpet_area.append(np.nan)
        # =============== Floor Number Extraction ===============
    floor_pattern = r"Floor\s*number\s*([0-9]{1,3})\s*of\s*([0-9]{1,3})\s*floors?"
    
    floor = re.findall(floor_pattern, text_block, flags=re.IGNORECASE)
    
    if floor:
        Floor_number.append(floor[0][0])      # Extract first number
        Total_floors.append(floor[0][1])      # Extract second number
    else:
        # Fallback for cases like "Floor 3 of 24"
        floor_pattern2 = r"Floor\s*([0-9]{1,3})\s*of\s*([0-9]{1,3})"
        floor2 = re.findall(floor_pattern2, text_block, flags=re.IGNORECASE)
        
        if floor2:
            Floor_number.append(floor2[0][0])
            Total_floors.append(floor2[0][1])
        else:
            Floor_number.append(np.nan)
            Total_floors.append(np.nan)
    
      # ------------ Bathrooms ------------
    bath = re.findall(r"Bathrooms?\s*(\d+)", text_block, flags=re.IGNORECASE)
    Bathrooms.append(bath[0] if bath else np.nan)
    
    # ------------ Balcony ------------
    balc = re.findall(r"Balcon(y|ies)\s*(\d+)", text_block, flags=re.IGNORECASE)
    Balcony.append(balc[0][1] if balc else np.nan)
    
    # ------------ Gas Pipeline ------------
    gas = re.findall(r"Gas Pipeline\s*(Yes|No)", text_block, flags=re.IGNORECASE)
    Gas_Pipeline.append(gas[0].title() if gas else np.nan)
    
    # ------------ Gated Community ------------
    gate = re.findall(r"Gate Community\s*(Yes|No)", text_block, flags=re.IGNORECASE)
    Gate_Community.append(gate[0].title() if gate else np.nan)
   
data={"Title":Title,"Locality":Locality,"Furnishing":Furnishing,"Built_up_area":Built_up_area,
      "Lease_type":Lease_type,"Available_from":Available_from,"Price":Price,"Bathrooms":Bathrooms,
      "Balcony":Balcony,"Carpet_area":Carpet_area,"Gas_Pipeline":Gas_Pipeline,"Gate_Community":Gate_Community,
      "Floor_number":Floor_number,"Total_floors":Total_floors}
df=pd.DataFrame(data)     
driver.quit()
```
